## Root Mean Squared Error - RMSE
<details>
    <summary>
        <a class="btnfire small stroke"><em class="fas fa-chevron-circle-down"></em>&nbsp;&nbsp;Clique para mais detalhes</a>
    </summary>
    <br>

A métrica Root Mean Squared Error é a raiz quadrada do erro quadrático médio, ou seja, é a raiz quadrada do Mean Squared Error. A biblioteca Scikit Learn não tem uma função implementatada para calcular o RMSE, mas é possível utilizar a função mean_squared_error para calcular o o MSE e depois utilizar o NumPy para calcular o RSME. <br>
A fórmula do Root Mean Squared Error é
$\sqrt{MSE}$
    
</details>

## $R^2$ - Coeficiente de determinação
<details>
    <summary>
        <a class="btnfire small stroke"><em class="fas fa-chevron-circle-down"></em>&nbsp;&nbsp;Clique para mais detalhes</a>
    </summary>
    <br>
$R^2$ é o coeficiente de determinação. Também é chamado de R score, ou R squared, ou ainda, R ao quadrado. <br>
Seu valor varia de 0 a 1. Sendo que, quanto mais próximo de um, melhor, e quanto mais próximo de zero pior. A escala de zero a um pode ser representada em valores percentuais. <br>
É uma metrica que avalia os acertos.
</details>

## Coletando os dados

```python
# Coletando os dados
df = pd.read_csv('Dados/Auto-mpg/mtcars.csv')
df
```

```python
# Visualizando informações sobre o dataset
df.info()
```

```python
# Separando as variáveis preditoras e a variável alvo
numeroObservacoes = len(df)
numeroVariaveisPreditoras = 10
X = df[["cyl","disp","hp","drat","wt","qsec","vs","am","gear","carb"]].values.reshape((numeroObservacoes, numeroVariaveisPreditoras)) # X deve sempre ser uma matriz e nunca um vetor
y = df['mpg'].values # y pode ser um vetor

# Cria o objeto SelectKBest
selectkBest = SelectKBest(score_func = f_regression, k=4)

# Executa a função em (X, y) e obtém as variáveis selecionadas
selectkBestTreinado = selectkBest.fit(X, y)

# Reduz X para as variáveis selecionadas
variaveisSelecionadas = selectkBestTreinado.transform(X)

# Cria uma lista com o nome das variáveis
variaveis = ["cyl","disp","hp","drat","wt","qsec","vs","am","gear","carb"]

# Resultados
print('\nRanking:', selectkBestTreinado.scores_)
plt.bar(variaveis, selectkBestTreinado.scores_)
```

```python
# Separando as variáveis preditoras e a variável alvo
numeroObservacoes = len(df)
numeroVariaveisPreditoras = 10
X = df[["cyl","disp","hp","drat","wt","qsec","vs","am","gear","carb"]].values.reshape((numeroObservacoes, numeroVariaveisPreditoras)) # X deve sempre ser uma matriz e nunca um vetor
y = df['mpg'].values # y pode ser um vetor
```

```python
# Divide os dados em treino e teste
Xtreino, Xteste, Ytreino, Yteste = train_test_split(X, y, test_size = 0.2, random_state=23)
```

```python
# Criando o modelo
modelo = LinearRegression()

# Treinando o modelo
modelo.fit(Xtreino, Ytreino)

# Fazendo previsões
previsoes = modelo.predict(Xteste)

# r2
r2 = r2_score(Yteste, previsoes)

# MSE
mse = mean_squared_error(Yteste, previsoes)

# RMSE
rmse = np.sqrt(mse)

# MAE
mae = mean_absolute_error(Yteste, previsoes)

print("R2 do modelo:", r2*100)
print("MSE do modelo:", mse)
print("RMSE do modelo:", rmse)
print("MAE do modelo:", mae)
```

```python
dicionario = {
    "Yteste": Yteste,
    "Valor Previsto": previsoes
}

dfComparacao = pd.DataFrame(dicionario)
dfComparacao
```

```python
# Criando o modelo
modelo = DecisionTreeRegressor(random_state = 35)

# Treinando o modelo
modelo.fit(Xtreino, Ytreino)

# Fazendo previsões
previsoes = modelo.predict(Xteste)

# r2
r2 = r2_score(Yteste, previsoes)

# MSE
mse = mean_squared_error(Yteste, previsoes)

# RMSE
rmse = np.sqrt(mse)

# MAE
mae = mean_absolute_error(Yteste, previsoes)

print("R2 do modelo:", r2*100)
print("MSE do modelo:", mse)
print("RMSE do modelo:", rmse)
print("MAE do modelo:", mae)
```

```python
dicionario = {
    "Yteste": Yteste,
    "Valor Previsto": previsoes
}

dfComparacao = pd.DataFrame(dicionario)
dfComparacao
```

```python
# Criando o modelo
modelo = SVR()

# Treinando o modelo
modelo.fit(Xtreino, Ytreino)

# Fazendo previsões
previsoes = modelo.predict(Xteste)

# r2
r2 = r2_score(Yteste, previsoes)

# MSE
mse = mean_squared_error(Yteste, previsoes)

# RMSE
rmse = np.sqrt(mse)

# MAE
mae = mean_absolute_error(Yteste, previsoes)

print("R2 do modelo:", r2*100)
print("MSE do modelo:", mse)
print("RMSE do modelo:", rmse)
print("MAE do modelo:", mae)
```
