## Similaridade por cosseno

A **similaridade por cosseno** é uma medida de similaridade entre dois vetores diferentes de zero e é calculada como o ângulo entre esses vetores, de tal forma, que quanto menor o ângulo entre dois vetores, mais similares eles são. Por outro lado, quanto maior o ângulo entre dois vetores, menor é a similaridade entre ambos. <br>
O cálculo da similaridade por cosseno envolve o uso do cosseno do ângulo entre dois vetores. A fórmula é dada por:<br> <br>
similaridade = cos(θ) = (A • B) / (||A|| * ||B||) <br><br>
Nessa fórmula, A e B representam dois vetores. O termo (A • B) representa o produto escalar entre os vetores A e B, enquanto ||A|| e ||B|| representam as normas (ou magnitudes) dos vetores A e B, respectivamente. <br>
O valor resultante do cálculo da similaridade por cosseno varia entre -1 e 1. Quanto mais próximo de 1, maior é a similaridade entre os vetores, indicando que eles apontam em uma direção semelhante no espaço vetorial. Um valor de -1 indica uma dissimilaridade máxima, ou seja, os vetores são opostos entre si. Um valor próximo a 0 indica que os vetores são ortogonais, ou seja, não possuem nenhuma similaridade. <br>
Em Ciência de Dados, cada linha de um conjunto de dados é um vetor. <br>
A similaridade por cosseno também é frequentemente usada para comparar a similaridade entre documentos de texto. Nesse caso, os documentos são representados como vetores em um espaço vetorial, onde cada dimensão pode ser um termo ou uma característica específica. A similaridade por cosseno permite encontrar documentos que são semanticamente semelhantes ou relacionados com base nas palavras ou características compartilhadas. <br>
Em resumo, a similaridade por cosseno é uma métrica que quantifica a semelhança entre dois vetores com base no cosseno do ângulo entre eles, permitindo avaliar a proximidade ou relação entre eles em um espaço vetorial. Os vetores podem ser as linhas de um conjunto de dados ou documentos em texto. <br>
O cálculo da similaridade por cosseno exige que os dados estejam normalizados.

```python
# Carregando os pacotes
import pandas as pd
import numpy as np

# Normalização dos dados
from sklearn.preprocessing import MinMaxScaler

# Calcula a similiraridade de cosseno
from sklearn.metrics.pairwise import cosine_similarity

# Algoritmo KMeans
from sklearn.cluster import KMeans
```

```python
# Coletando os dados 
df = pd.read_csv('Dados/Clientes/dados_clientes.csv')
df
```

```python
# Excluindo coluna
df = df.drop(["IDCliente"], axis=1)
df
```

## Normalizando os dados

```python
# Retorna os valores do DataFrame e converte em uma matriz NumPy
valores = df[["Idade", "Salario_Mensal_Milhar", "Pontuacao_Gasto"]].values

# Cria o objeto da classe MinMaxScaler 
min_max_scaler = MinMaxScaler()

# Realiza a normalização dimensionando as variáveis em uma escala entre 0 e 1
dadosNormalizados = min_max_scaler.fit_transform(valores)
dadosNormalizados
```

## Calculando a similaridade de cosseno

```python
# Calacula a similaridade de cosseno
similaridade = cosine_similarity(dadosNormalizados)
similaridade
```

```python
# Visualizando o tipo de dado do objeto
type(similaridade)
```

```python
# Visualizando os valores mínimo e máximo das similaridades
print("Valor mínimo:", np.min(similaridade))
print("Valor máximo:", np.max(similaridade))
```

```python
# Criando o objeto 
kmeans = KMeans(n_clusters=6 , random_state=0)

# Treinando o modelo e fazendo as previsões dos clusters
kmeans.fit_predict(similaridade)

# Visualizando o cluster atribuído para cada observação
kmeans.labels_
```

```python
# Associando cada cliente a seu respectivo grupo
dfKmeans = df.copy()
dfKmeans["Cluster"] = kmeans.labels_
dfKmeans
```

```python
# Selecionando um cliente aleatório
cliente1 = dfKmeans.sample(random_state=7)
cliente1
```

```python
# Pesquisando clientes semelhantes
dfClientesSimilares = dfKmeans[dfKmeans.Cluster == int(cliente1.Cluster)]
dfClientesSimilares
```
