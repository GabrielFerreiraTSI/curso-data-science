Muitos conjuntos de dados do mundo real contêm valores ausentes, geralmente codificados como espaços em branco, NA ou NaNs, também chamados de valores nulos, ou ainda missing values em inglês. Esses valores nulos, no entanto, são incompatíveis com os algoritmos de aprendizado de máquina, que esperam que todos os valores em uma matriz sejam numéricos e que todos tenham um significado. Isso significa, que os valores nulos precisam passar por um tratamento. Ou seja, o cientista de dados não pode entregar valores ausentes aos algoritmos de Machine Learning. <br>
Uma estratégia simples para tratar valores ausentes é excluir linhas e/ou colunas inteiras contendo esses valores. No entanto, a desvantagem dessa estratégia é que, ao excluir linhas ou colunas, podemos perder dados que podem ser valiosos, mesmo que incompletos, principalmente quando o conjunto de dados é pequeno. <br>
Existem duas técnicas de imputação dos dados: <br>
- A técnica **univariada** considera apenas os valores conhecidos de uma coluna para imputar os valores nulos. Com a técnica univariada os valores nulos podem ser imputados com um valor constante fornecido pelo cientista de dados ou usando as estatísticas como média, mediana ou a moda, que é o valor mais frequente de cada coluna na qual os valores NA estão localizados. <br>
- A técnica **multivariada** é mais sofisticada, pois modela cada coluna do dataset com valores ausentes em função de outras colunas e usa essa estimativa para imputação. A técnica multivariada age de forma iterativa, onde em cada etapa, uma coluna é designada como saída y e as outras colunas são tratadas como entradas X. Dessa forma, um algoritmo de regressão Randon Forest é treinado em (X, y) para y conhecido. Em seguida, o modelo de regressão é usado para prever os valores ausentes de y.

### Técnicas univariadas utilizando o Pandas

```python
# A coluna Cabin pode ser excluída, pois possui muitos valores nulos
df2 = df.drop("Cabin", axis=1)
df2
```

```python
# Retornando somente as linhas com valores nulos
df2[df2.isnull().any(axis=1)]
```

#### Imputação usando valores constantes

```python
# Faz uma cópia do dataset
df3 = df2.copy()
```

```python
# A função fillna substitui os valores nulos pelo valor desejado
# Caso deseje utilizar a função em todo dataset, utilize df = df.fillna(n) 

# Define o valor que desejamos inserir
n = 99

# Substitui todos os valores nulos da coluna pelo valor de n
df3.Age = df3.Age.fillna(n)
```

```python
# Retorna a quantidade de valores nulos por coluna
df3.isnull().sum() 
```

```python
# Pesquisando uma idade específica
df3[df3.Age == 99]
```

#### Imputação usando as estatísticas como média, mediana ou a moda

```python
# Verificando a média de idade
df2.Age.mean()
```

```python
# Faz uma cópia do dataset
df4 = df2.copy()
```

```python
# Substitui todos os valores NA pela média da respectiva coluna
# Caso deseje utilizar a função em todo dataset, utilize df = df.fillna(df.mean())

df4.Age = df4.Age.fillna(df4.Age.mean())
```

```python
# Pesquisando uma idade específica
df4[df4.Age == df2.Age.mean()]
```

```python
# Verificando a mediana de idade
df2.Age.median()
```

```python
# Faz uma cópia do dataset
df5 = df2.copy()
```

```python
# Substitui todos os valores NA pela mediana da respectiva coluna
# Caso deseje utilizar a função em todo dataset, utilize df = df.fillna(df.mean())

df5.Age = df5.Age.fillna(df5.Age.median())
```

```python
# Pesquisando uma idade específica
df5[df5.Age == df2.Age.median()]
```

```python
# Visualizando a moda
moda = df2.Age.mode()
moda
```

```python
# Faz uma cópia do dataset
df6 = df2.copy()
```

```python
# Substitui todos os valores NA pela moda da respectiva coluna
# O método mode do Pandas retorna uma Series, por isso devemos informar a posição que desejamos da Series.
df6.Age = df6.Age.fillna(moda[0])
```

```python
# Pesquisando uma idade específica
df6[df6.Age == moda[0]]
```

### Técnicas univariadas utilizando o Scikit-Learn

#### Imputação usando valores constantes

```python
# Faz uma cópia do dataset
df7 = df2.copy()
```

```python
# Cria o objeto do tipo SimpleImputer com o valor constante como estratégia  
imputeConstante = SimpleImputer(missing_values=np.nan, strategy="constant", fill_value = 99)
```

```python
# Treina o objeto  do tipo SimpleImputer, substitui os valores nulos pela constante e atribui à coluna Age
df7.Age = imputeConstante.fit_transform(df7.Age.values.reshape(-1,1))
```

```python
# Pesquisando uma idade específica
df7[df7.Age == 99]
```

#### Imputação usando as estatísticas como média, mediana ou a moda

```python
# Faz uma cópia do dataset
df8 = df2.copy()
```

```python
# Cria o objeto do tipo SimpleImputer com a média como estratégia  
imputeMedia = SimpleImputer(missing_values=np.nan, strategy= "mean")
```

```python
# Treina o objeto  do tipo SimpleImputer, substitui os valores nulos pela média e atribui à coluna Age
df8.Age = imputeMedia.fit_transform(df8.Age.values.reshape(-1,1))
```

```python
# Pesquisando uma idade específica
df8[df8.Age == df2.Age.mean()]
```

### Técnicas multivariadas utilizando o Scikit-Learn

```python
# Faz uma cópia do dataset
df9 = df2.copy()
```

```python
# Excluindo a coluna "PassengerId"
df9 = df9.drop("PassengerId", axis=1)
df9
```

```python
# Retorna as colunas numéricas e armazena na lista colunaNumerica

colunasNumericas = []

for coluna in df9.columns:
    if df9[coluna].dtype != object:
        colunasNumericas.append(coluna)
colunasNumericas
```

```python
# Imputação Iterativa
imputacaoInterativa = IterativeImputer(random_state=30)
valoresImputados = imputacaoInterativa.fit_transform(df[colunasNumericas])
dfValoresImputados = pd.DataFrame(valoresImputados, columns=colunasNumericas)
dfValoresImputados
```

```python
# Criando um DataFrame para visualizar os valores imputados
dfComparaIdade = pd.DataFrame(df.Age)
dfComparaIdade["Idade Imputada"] = dfValoresImputados.Age

# Retornando somente as linhas com valores nulos
dfComparaIdade[dfComparaIdade.isnull().any(axis=1)]
```

```python
# Visualizando os valores iniciais e os valores imputados através de boxplot
ax = sns.boxplot(data=[df.Age, dfComparaIdade["Idade Imputada"]], palette="Set2")
ax.set_xticklabels(["Idade inicial", "Idade Imputada"]) # exibe o nome da variável
plt.show()
```
