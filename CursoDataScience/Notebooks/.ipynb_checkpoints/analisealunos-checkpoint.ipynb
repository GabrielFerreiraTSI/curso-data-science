# <font color='blue'>Etapa 1 - Coletando os dados</font>

```python
import pandas as pd

# Leitura de arquivo EXCELL
dados = pd.read_excel("notas-ite-11-2017.xlsx", sheet_name="MÉDIAS FINAIS")
dados
```

```python
# Leitura de arquivo CSV
dadosDiversos = pd.read_csv('dados-estudantes.csv', sep = ';', encoding = 'latin-1')
dadosDiversos
```

# <font color='blue'>Etapa 2: Explorando e Preparando os Dados</font>

```python
# Excluir linhas
dados = dados.drop([0,1,2,3,4,5,39], axis=0)
dados
```

```python
dados.columns = ["Nome","historia","falta-historia","quimica","falta-quimica","sociologia","falta-sociologia","filosofia","falta-filosofia","portugues","falta-portugues","matematica","falta-matematica","fisica","falta-fisica","desenho","falta-desenho","ed-fisica","falta-ed-fisica","informatica","falta-informatica","materiais","falta-materiais","artes","falta-artes","total-faltas","presenca","situacao"]
#dados
```

```python
dados['Nome'] = dados['Nome'].str.upper()
#dados
```

```python
# Merge de Dataframes
left = dados
right = dadosDiversos
df = pd.merge(left, right, on='Nome')
```

```python
df.columns
```

```python
# countplot
import seaborn as sns
sns.set(rc={'figure.figsize':(11.7,6.27)})
ax = sns.countplot(x = "situacao", data = df, palette = "Greens_d");
ax.set_title("Resultado dos alunos no final do ano", fontsize = 16)
ax.set_xlabel("Situação")
ax.set_ylabel("Quantidade")
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + 0.35, p.get_height() + 0.2))
```

```python
# Frequência relativa
df2 = pd.value_counts(df["situacao"]) / len(df["situacao"])*100
df2
```

```python
# Quantidade de alunos por cidade
sns.set(rc={'figure.figsize':(11.7,6.27)})
ax = sns.countplot(x = "Cidade", data = df, palette = "Greens_d");
ax.set_title("Quantidade de alunos por cidade", fontsize = 16)
ax.set_xlabel("Cidade")
ax.set_ylabel("Quantidade")
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + 0.35, p.get_height() + 0.2))
```

```python
# Quantidade de alunos por Forma de Ingresso
sns.set(rc={'figure.figsize':(11.7,6.27)})
ax = sns.countplot(y = "Forma de Ingresso", data = df, palette = "Greens_d");
ax.set_title("Quantidade de alunos por Forma de Ingresso", fontsize = 16)
ax.set_xlabel("Quantidade")
ax.set_ylabel("Forma de Ingresso")
for p in ax.patches:
    _x = p.get_x() + p.get_width()
    _y = p.get_y() + p.get_height() - 0.4
    value = int(p.get_width())
    ax.text(_x, _y, value, ha="left")
```

```python
df3 = df3.drop(['falta-historia', 'falta-quimica',"falta-sociologia","falta-filosofia",'falta-portugues',"falta-matematica","falta-fisica","falta-desenho",
               "falta-ed-fisica","falta-informatica","falta-materiais","falta-artes","presenca","#","Matrícula","Endereço"], axis=1)
```

```python
dfSituacaoFormaIngresso = df3.groupby(["situacao", "Forma de Ingresso"])
dfSituacaoFormaIngresso.size()
```

```python
ax = dfSituacaoFormaIngresso.size().plot.barh()
ax.set_title("Relação do resultado do aluno e a Forma de Ingresso", fontsize = 16)
ax.set_xlabel("Quantidade")
ax.set_ylabel("Resultado/Forma de Ingresso")
for p in ax.patches:
    _x = p.get_x() + p.get_width()
    _y = p.get_y() + p.get_height() - 0.4
    value = int(p.get_width())
    ax.text(_x, _y, value, ha="left")
```
