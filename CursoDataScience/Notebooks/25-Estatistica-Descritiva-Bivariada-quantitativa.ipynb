### Gráfico de dispersão
<details>
    <summary>
        <a class="btnfire small stroke"><em class="fas fa-chevron-circle-down"></em>&nbsp;&nbsp;Clique para mais detalhes</a>
    </summary>
    <br>

A correlação entre duas variáveis quantitativas pode ser representada de forma gráfica por meio de um diagrama de dispersão. Este diagrama representa graficamente os valores das variáveis X e Y em um plano cartesiano. Um gráfico de dispersão permite, portanto, avaliar: <br> 

a) se existe ou não alguma correlação entre as variáveis em estudo; <br> 
b) o tipo de correlação entre as duas variáveis, isto é, a direção em que a variável Y aumenta ou diminui em função da variação de X; <br>
c) o grau de correlação entre as variáveis; <br>
d) a natureza da correlação (linear, exponencial, etc.). <br>

```python
# Importando os pacotes
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Coletando os dados
df = pd.read_csv('Dados/Auto-mpg/mtcars.csv')
df.head()
```

```python
# Gráfico de dispersão utilizando Seaborn com Modelo de Regressão Linear margem de erro
# Correlação positiva 
ax = sns.lmplot(x = "drat", y = "mpg", data = df)
plt.show()
```

```python
# Gráfico de dispersão utilizando Seaborn com Modelo de Regressão Lienar e margem de erro
# Correlação negativa
ax = sns.lmplot(x = "hp", y = "mpg", data = df)
plt.show()
```

```python
# Gráfico de dispersão utilizando Seaborn com Modelo de Regressão Lienar margem de erro
# Sem Correlação
ax = sns.lmplot(x = "gear", y = "hp", data = df)
plt.show()
```

### Coeficiente de correlação
<details>
    <summary>
        <a class="btnfire small stroke"><em class="fas fa-chevron-circle-down"></em>&nbsp;&nbsp;Clique para mais detalhes</a>
    </summary>
    <br>

O coeficiente de correlação é uma medida que varia entre -1 e 1. <br>
Por meio do sinal, é possível verificar o tipo de relação linear entre as duas variáveis analisadas, ou seja a direção em que a variável Y aumenta ou diminui em função da variação de X. <br>
Quanto mais próximo dos valores extremos, mais forte é a correlação entre elas. Logo: <br>
- Se o coeficiente de correlação for positivo, existe uma relação diretamente proporcional entre as variáveis; <br>
- se o coeficiente de correlação for igual a 1, tem-se uma correlação linear positiva perfeita. <br>
- Se o coeficiente de correlação for negativo, existe uma relação inversamente proporcional entre as variáveis; <br>  
- se o coeficiente de correlação for igual a –1, tem-se uma correlação linear negativa perfeita; <br>
- Se o coeficiente de correlação for igual a zero, não existe correlação. <br>

Existem três modelos matemáticos para calcular o coeficiente de correlação, que levam o respectivo nome do pesquisador que o escreveu. São eles: o método de Pearson, o método de Kendall e o método de Spearman.
</details>

```python
# Calculando o coeficiente de correlação entre todas as variáveis.
# Pearson é o modelo matemático padrão. Caso, deseje utilizar outro modelo matemático, basta informar como parâmetro 
# (Ex.: df.corr(numeric_only = True, method = "kendall") ou df.corr(numeric_only = True, method = "spearman"))
df.corr(numeric_only = True)
```

```python
# Matriz de correlação utilizando Matplotlib
plt.figure(figsize=(10, 8))
plt.matshow(df.corr(numeric_only = True))
plt.xticks(range(df.select_dtypes(['number']).shape[1]), df.select_dtypes(['number']).columns, fontsize=14, rotation=45)
plt.yticks(range(df.select_dtypes(['number']).shape[1]), df.select_dtypes(['number']).columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Matriz de correlação', fontsize=16);
plt.show()
```

```python
# Matriz de correlação utilizando Plotly
import plotly.express as px

# Matriz de correlação 
fig = px.imshow(df.corr(numeric_only = True), 
                title="Matriz de Correlação",
                aspect="auto",
                width=800, 
                height=400)
fig.update_xaxes(side="top") # exibe o nome das variáveis acima do gráfico
fig.show()
```

```python
# Matriz de correlação utilizando Seaborn
sns.set(rc = {'figure.figsize':(10,8)}) # define o tamanho do gráfico
sns.heatmap(df.corr(numeric_only = True),
            annot = True,
            fmt = '.2f',
            cmap='Blues')
plt.title('Matriz de correlação', fontsize = 20)
plt.show()
```

```python
# Importando os pacotes
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```
